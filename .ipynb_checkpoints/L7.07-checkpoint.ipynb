```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
# Apply the Random Forests algorithm but this time only by upscaling the data
numerical = pd.read_csv('files_for_lab/numerical.csv')
categorical = pd.read_csv('files_for_lab/categorical.csv')
targets = pd.read_csv('files_for_lab/target.csv')
data = pd.concat([numerical, categorical, targets], axis = 1)
```

```python
data.head()
```

```python
data.shape
```

```python
data.isna().sum().sum()
```

```python
y = data['TARGET_B']
X = data.drop(['TARGET_B'], axis = 1)

numericalX = X.select_dtypes(np.number)
categoricalX = X.select_dtypes(np.object)

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first').fit(categoricalX)
encoded_categorical = encoder.transform(categoricalX).toarray()
encoded_categorical = pd.DataFrame(encoded_categorical)
X = pd.concat([numericalX, encoded_categorical], axis = 1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# for upsampling we need to temporarily concat X_train and y_train
trainset = pd.concat([X_train, y_train], axis=1)

from sklearn.utils import resample
category_0 = trainset[trainset['TARGET_B']== 0 ]
category_1 = trainset[trainset['TARGET_B']== 1 ]
category_1 = resample(category_1,
                    replace=True, 
                    n_samples = len(category_0),
                    random_state=0)
```

```python
display(category_0.shape)
display(category_1.shape)
```

```python
trainset_new = pd.concat([category_0, category_1], axis = 0)
trainset_new = trainset_new.sample(frac =1) #randomize the rows
X_train = trainset_new.drop(['TARGET_B'], axis=1)
y_train = trainset_new['TARGET_B']
```

```python
X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)

y_train_regression = X_train['TARGET_D']
y_test_regression = X_test['TARGET_D']

X_train = X_train.drop(['TARGET_D'], axis = 1)
X_test = X_test.drop(['TARGET_D'], axis = 1)
```

```python
y_test.value_counts()
```

```python
y_train.value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

clf = RandomForestClassifier(max_depth=5,
                             min_samples_split=20,
                             min_samples_leaf =20,
                             max_samples=0.2,
                             random_state = 42)
clf.fit(X_train, y_train)
print(clf.score(X_train, y_train))
print(clf.score(X_test, y_test))

y_pred = clf.predict(X_test)
display(y_test.value_counts())
display(confusion_matrix(y_test, y_pred))
```

```python
# For cross validation (it takes several minutes to run!!)
from sklearn.model_selection import cross_val_score
clf = RandomForestClassifier(max_depth=5,
                             min_samples_split=20,
                             min_samples_leaf =20)
cross_val_scores = cross_val_score(clf, X_train, y_train, cv=10)
print(np.mean(cross_val_scores))
```

```python
# Discuss the output and its impact in the bussiness scenario
# I assume each mail has a cost of $0.68 (as stated in class). The cost of a false positive is $0.68 
# and there is no gift, so we lose $0.68 for each false positive.
# On the other hand, a false negative implies we are losing a gift (average gift of $15.62) taking into 
# account only the donors). This false prediction has a worse impact in the return. Each false negative
# has the same impact as 23 false positives
# In order to maximize the return, we should try to improve accuracy when predicting positives (minimizing 
# false negatives), even if this implies a loss of accuracy predicting negatives. 
# By introducing imbalance to positives in data (by instance, upsamplig with more than 50% of total data) 
# the model detects more positives, at the cost of more false positives. I have tried it and seems to work 
# fine, with some improvement in net return.   
# With downsamplig (class). Return for test = 590*15.76-(7751+590)*0.68 = 3543.92
# With upsamplig (lab). Return for test = 542*15.76-(6911+542)*0.68 = 3473.88
# Upsampling seems to perform a little better, since it predicts better the positives, but predictions vary slightly every
# time we execute the model.
```
